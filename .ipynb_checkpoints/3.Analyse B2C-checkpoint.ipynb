```python
# prérequis pour utiliser resample() : formatage en datetime + indexation de la colonne 'date'
serie_temp = lapage.copy()
serie_temp['date'] = serie_temp['date'].apply(pd.to_datetime)
serie_temp.set_index('date', inplace=True)
serie_temp = serie_temp[['price']].rename(columns={'price': 'CA'})
```

#### *CA hebdomadaire*

```python
w = serie_temp.resample(rule='W').sum()
# imputation des valeurs pour octobre pour une meilleure lisibilité du graphique et de la tendance globale
list = ['2021-10-03', '2021-10-10', '2021-10-17', '2021-10-24', '2021-10-31'] 
for i in list:
    w.loc[i] = w.iloc[29:36:6].mean() #imputation par la moyenne des 2 semaines avant et après octobre 2021
    
# imputation des valeurs pour les mois aux extrémités du timeseries pour 
                #une meilleure lisibilité du graphique et de la tendance globale
w.iloc[104] = w.iloc[103]
w.tail(3)
```

#### *CA mensuel*

```python
m = serie_temp.resample(rule='M').sum()
# imputation des données d'octobre 2021 par la moyenne de septembre et novembre 2021
m.iloc[7] = m.iloc[6:9:2].mean()
m[6:9]
```

#### *CA trimestriel*

```python
q = m.resample(rule='Q').sum() #resample sur m car les données mensuelles d'octobre ont été traitées ???
# imputation des valeurs pour les mois aux extrémités du timeseries pour 
                #une meilleure lisibilité du graphique et de la tendance globale
q.loc['2021-03-31', 'CA'] = q['CA'][0] * 3 #car on 1/3 des données
q.loc['2023-03-31', 'CA'] = q['CA'][8] * 3/2 #car on 2/3 des données
q
```

## ***Profils clients*** (répartition CA entre eux) - ***courbe de Lorenz***

```python
# mesure de concentration (courbes de Lorenz et indice de Gini)
depenses = lapage[lapage['price'] > 0]
dep = depenses['price'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()

# la courbe de Lorenz commence à 0
lorenz = np.append([0],lorenz)

# surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
AUC = (lorenz.sum() - lorenz[-1] / 2 - lorenz[0] / 2) / n

# surface entre la première bissectrice et le courbe de Lorenz
S = 0.5 - AUC 
gini = 2*S
print(bold + red + f"\nL'indice de Gini est égal à {round(float(gini), 2)}, la répartition des clients par rapport aux CA n'est pas égalitaire")

# il y a un segment de taille n pour chaque individu, plus 1 segment supplémentaire d'ordonnée 0. Le premier segment commence à 0-1/n, et le dernier termine à 1+1/n.
xaxis = np.linspace(0-1/n,1+1/n,n+1)

# visualisation
fig, ax = plt.subplots(figsize=(6, 6))

plt.plot(xaxis,lorenz,drawstyle='steps-post', label = 'courbe de Lorenz')
plt.plot(xaxis,xaxis, label ='égalité parfaite')

plt.axhline(y=0.5, xmin=0, xmax=0.74, c='red', lw=1, linestyle='dashed',label ='médiale')
plt.axvline(x=0.76, ymin=0, ymax=0.5, c='red', lw=1, linestyle='dashed')
ax.text(-0.12, 0.49, "0.5",color='red')
ax.text(0.69, -0.10, "0.76",color='red')

plt.axhline(y=0.25, xmin=0, xmax=0.5, c='green', lw=1, linestyle='dashed',label ='médiane')
plt.axvline(x=0.5, ymin=0, ymax=0.27, c='green', lw=1, linestyle='dashed')
ax.text(-0.14, 0.24, "0.25",color='green')
ax.text(0.47, -0.10, "0.5",color='green')

plt.xlabel("part cumulée des clients (%)")
plt.ylabel("part cumulée du CA (%)")
plt.legend()
plt.title('Répartition du CA entre les clients\n', fontsize=14)
plt.show()
```

<span style="color:red">**=> 50% des clients génèrent 75% du CA et 50% du CA est généré par 24% des clients**</span>


### Comportement des clients

```python
# utilisation d'un pivot table
# CA et nombre d'achats par client
client = lapage.pivot_table(values=['session_id', 'price'], index=['client_id', 'age', 'sex'],
                    aggfunc={'session_id': len,
                             'price': np.sum}).rename(columns = {'session_id': 'nb_tran', 'price': 'CA'})
client['ratio CA / nb_tran = panier moyen'] = client['CA'] / client['nb_tran']
client = client.sort_values(by='nb_tran', ascending=False)
client.head()
```

#### Vérification de la distribution de la variable 'age'

Cette vérification nous servira par la suite pour la réalisation des tests de corrélation.

<span style="color:red">**Avec le QQ plot :**</span>

```python
# QQ Plot
from numpy.random import seed
from numpy.random import randn
from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot
# seed the random number generator
#seed(1)
# generate univariate observations
#data = 5 * randn(100) + 50
# q-q plot
qqplot(lapage['age'], line='s')
pyplot.show()
```

```python
shapiro_test = stats.shapiro(lapage['age'])
shapiro_test
```

<span style="color:red">**Étant donné que la p-value calculée est inférieure au seuil de signification alpha = 0,05, on rejette l'hypothèse nulle H0.**</span>

### Test d'indépendance Chi-2 (variables quali-quali)

Considérons maintenant la variable 'categ' comme de type catégorique (qualitative)

<span style="color:red">**Nous allons vérifier avec le test du Chi-2 l'indépendance des deux variables "sex" et "catégorie" du client :**</span>

<span style="color:red">**Tout d'abord vérifions les conditions pour utiliser ce test :**</span>

+ 1. Les mesures sont indépendantes
+ 2. Pas d'effectifs théoriques inférieurs à 5 dans les cases du tableau de contigence

<span style="color:red">**Interprétation du test :**</span> 
+ H0 : Les deux variables qualitatives sont indépendantes
+ H1 : Les deux variables qualitatives dépendent l’une de l’autre

#### Tableau de contingence 

```python
# changement du format de 'categ' en variable qualitative
lapage['categ'] = lapage['categ'].astype('object')
# vérification du format quali/quali
lapage[['categ', 'sex']].dtypes
```

```python
# création du tableau de contingence "Matrice des valeurs observées"
X = "sex"  
Y = "categ"
cont = lapage[[X,Y]].pivot_table(index=X, columns=Y, aggfunc=len, margins=True, margins_name="Total")
cont
```

```python
# heatmap du tableau de contingence
tx = cont.loc[:,["Total"]]
ty = cont.loc[["Total"],:]
n = len(cont)
indep = tx.dot(ty) / n

c = cont.fillna(0) # on remplace les valeurs nulles par 0
measure = (c-indep)**2/indep
xi_n = measure.sum().sum()
table = measure/xi_n
sns.heatmap(table.iloc[:-1,:-1],annot=c.iloc[:-1,:-1])
plt.show()
```

#### Hypothèses

<span style="color:red">**Alpha = 0.01**</span>

<span style="color:red">**H0 : le genre n'influence pas le choix de la catégorie**</span>

<span style="color:red">**H1 : le genre influence le choix de la catégorie de produits**</span>

#### Calcul du chi-2 et p_value

```python
stats_chi2, stats_p, stats_dof, stats_exp = stats.chi2_contingency(cont)
p_value = stats_p
Chi2 = stats_chi2
DegréDeLiberte = stats_dof
print('\nchi-2 =',Chi2)
print('degré de liberté =',DegréDeLiberte)
print('p_value =',p_value)
```

<span style="color:red">**Étant donné que la p-value calculée est inférieure au seuil de signification alpha = 0,01, on rejette l'hypothèse nulle H0.**</span>

<span style="color:red">**=> Le genre et la catégorie dépendent l’une de l’autre**</span>

## ***Lien entre âge et montant total des achats***

### Test de corrélation de Spearman (variables quanti-quanti)

<span style="color:red">**La distribution de la variable âge n'étant pas normale, nous allons utiliser le test de corrélation Spearman pour vérifier le lien entre "age" et "montant total des achats" :**</span>

<span style="color:red">**Tout d'abord vérifions les conditions pour utiliser ce test :**</span>

+ 1. Test non-paramétrique, s'utilise en cas de distribution non normale
+ 2. Peut s'utiliser avec des données ordinales ou discrètes contrairement au test Pearson (utilise que des données continues)
+ 3. Les observations dans chaque échantillon sont indépendantes et identiquement distribuées (iid) 
+ 4. Les observations dans chaque échantillon peuvent être classées par rang

<span style="color:red">**Interprétation de la force de corrélation :**</span>

|\|ρ\|| Interpretation|
|-------|---------------|
|0.00 < 0.10| Negligible|
|0.10 < 0.20 |Weak|
|0.20 < 0.40| Moderate|
|0.40 < 0.60| Relatively strong|
|0.60 < 0.80| Strong|
|0.80 <= 1.00| Very strong|

<span style="color:red">**Cependant, cela ne montre pas la significativité du lien, pour cela nous allons utliser le test de Spearman :**</span>

<span style="color:red">**Interprétation du test :**</span> 
+ H0 : La corrélation n'est pas significative : ρ = 0
+ H1 : La corrélation est significative : ρ ≠ 0

```python
# création d'une variable 'age_price' avec le CA en fonction de l'âge des clients
age_price = lapage.groupby('age').sum().reset_index()
age_price = age_price[['age', 'price']].sort_values(by='age', ascending=False) 
age_price.head()
```

```python
# visualisation avec un scatterplot
sns.set_style('ticks')
sns.set_context('talk')
fig, ax = plt.subplots(figsize=(15, 8))
plt.plot(age_price['age'], age_price['price'], 'o', color='red')
sns.regplot(age_price['age'], age_price['price'])

plt.xlabel('Age')
plt.ylabel('Montant achat (€)')
plt.title('Montant total des achats selon l\'âge du client', fontsize=20)

plt.show()
```

```python
# heatmap : corrélation de Spearman
sns.heatmap(age_price.corr(method='spearman'), annot=True, fmt=".2f")
plt.title("Corrélation sur l'ensemble des variables quantitatives")
plt.show()
```

```python
# calcul du coefficient de corrélation linéaire de Spearman et p-value
coef_age_price = round(stats.spearmanr(age_price.age, age_price.price)[0], 2)
pvalue_age_price = stats.spearmanr(age_price.age, age_price.price)[1]
print(f"\nLe coefficient de corrélation Spearman est de {coef_age_price}\nLa p-value est de {pvalue_age_price}")
```

<span style="color:red">**Étant donné que la p-value calculée est inférieure au seuil de signification alpha = 0,05, on rejette l'hypothèse nulle H0.**</span>

<span style="color:red">**=> La corrélation est donc significative, les deux variables évoluent inversement avec une force de -0.87**</span>

### Test d'indépendance Chi-2 (variables quali-quali)

<span style="color:red">**Nous allons vérifier avec le test du Chi-2 l'indépendance des deux variables "classes d'âge" et "catégorie" du client :**</span>

<span style="color:red">**Tout d'abord vérifions les conditions pour utiliser ce test :**</span>

+ 1. Les mesures sont indépendantes
+ 2. Pas d'effectifs théoriques inférieurs à 5 dans les cases du tableau de contigence

<span style="color:red">**Interprétation du test :**</span> 
+ H0 : Les deux variables qualitatives sont indépendantes
+ H1 : Les deux variables qualitatives dépendent l’une de l’autre

#### Discrétisation de la variable 'age'

```python
# création de 4 classes d'âges
cut_bins = [17,30,45,60,94]
cut_labels = ["18-30","30-45", "45-60", "60+"]
lapage['class_age']= pd.cut(lapage['age'], bins=cut_bins, labels=cut_labels)
lapage.sample()
```

#### Tableau de contingence 

```python
# vérification du format quali/quali
lapage[['categ', 'class_age']].dtypes
```

```python
# création du tableau de contingence "Matrice des valeurs observées"
X = "class_age"  
Y = "categ"
cont = lapage[[X,Y]].pivot_table(index=X, columns=Y, aggfunc=len, margins=True, margins_name="Total")
cont
```

```python
# heatmap du tableau de contingence
tx = cont.loc[:,["Total"]]
ty = cont.loc[["Total"],:]
n = len(cont)
indep = tx.dot(ty) / n

c = cont.fillna(0) # on remplace les valeurs nulles par 0
measure = (c-indep)**2/indep
xi_n = measure.sum().sum()
table = measure/xi_n
sns.heatmap(table.iloc[:-1,:-1],annot=c.iloc[:-1,:-1])
plt.show()
```

#### Hypothèses

<span style="color:red">**Alpha = 0.01**</span>

<span style="color:red">**H0 : le genre n'influence pas le choix de la catégorie**</span>

<span style="color:red">**H1 : le genre influence le choix de la catégorie de produits**</span>

#### Calcul du chi-2 et p_value

```python
stats_chi2, stats_p, stats_dof, stats_exp = stats.chi2_contingency(cont)
p_value = stats_p
Chi2 = stats_chi2
DegréDeLiberte = stats_dof
print('\nchi-2 =',Chi2)
print('degré de liberté =',DegréDeLiberte)
print('p_value =',p_value)
```

<span style="color:red">**Étant donné que la p-value calculée est inférieure au seuil de signification alpha = 0,01, on rejette l'hypothèse nulle H0.**</span>

<span style="color:red">**=> L'âge et la catégorie dépendent bien l’une de l’autre**</span>
