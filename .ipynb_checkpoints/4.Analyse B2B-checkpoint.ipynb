```python
# prérequis pour utiliser resample() : formatage en datetime + indexation de la colonne 'date'
serie_temp = lapage.copy()
serie_temp['date'] = serie_temp['date'].apply(pd.to_datetime)
serie_temp.set_index('date', inplace=True)
serie_temp = serie_temp[['price']].rename(columns={'price': 'CA'})
```

#### *CA hebdomadaire*

```python
w = serie_temp.resample(rule='W').sum()
# imputation des valeurs pour octobre pour une meilleure lisibilité du graphique et de la tendance globale
list = ['2021-10-03', '2021-10-10', '2021-10-17', '2021-10-24', '2021-10-31'] 
for i in list:
    w.loc[i] = w.iloc[29:36:6].mean() #imputation par la moyenne des 2 semaines avant et après octobre 2021
    
# imputation des valeurs pour les mois aux extrémités du timeseries pour 
                #une meilleure lisibilité du graphique et de la tendance globale
w.iloc[104] = w.iloc[103]
w.tail(3)
```

#### *CA mensuel*

```python
m = serie_temp.resample(rule='M').sum()
# imputation des données d'octobre 2021 par la moyenne de septembre et novembre 2021
m.iloc[7] = m.iloc[6:9:2].mean()
m[6:9]
```

#### *CA trimestriel*

```python
q = m.resample(rule='Q').sum() #resample sur m
# imputation des valeurs pour les mois aux extrémités du timeseries pour 
                #une meilleure lisibilité du graphique et de la tendance globale
q.loc['2021-03-31', 'CA'] = q['CA'][0] * 3 #car on 1/3 des données
q.loc['2023-03-31', 'CA'] = q['CA'][8] * 3/2 #car on 2/3 des données
q
```

## ***Profils clients*** (répartition CA entre eux) - ***courbe de Lorenz***

```python
# mesure de concentration (courbes de Lorenz et indice de Gini)
depenses = lapage[lapage['price'] > 0]
dep = depenses['price'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()

# la courbe de Lorenz commence à 0
lorenz = np.append([0],lorenz)

# surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
AUC = (lorenz.sum() - lorenz[-1] / 2 - lorenz[0] / 2) / n

# surface entre la première bissectrice et le courbe de Lorenz
S = 0.5 - AUC 
gini = 2*S
print(bold + red + f"\nL'indice de Gini est égal à {round(float(gini), 2)}, la répartition des clients par rapport aux CA n'est pas égalitaire")

# il y a un segment de taille n pour chaque individu, plus 1 segment supplémentaire d'ordonnée 0. Le premier segment commence à 0-1/n, et le dernier termine à 1+1/n.
xaxis = np.linspace(0-1/n,1+1/n,n+1)

fig, ax = plt.subplots(figsize=(6, 6))

plt.plot(xaxis,lorenz,drawstyle='steps-post', label = 'courbe de Lorenz')
plt.plot(xaxis,xaxis, label ='égalité parfaite')

plt.axhline(y=0.5, xmin=0, xmax=0.765, c='red', lw=1, linestyle='dashed',label ='médiale')
plt.axvline(x=0.79, ymin=0, ymax=0.5, c='red', lw=1, linestyle='dashed')
ax.text(-0.12, 0.49, "0.5",color='red')
ax.text(0.7, -0.10, "0.79",color='red')

plt.axhline(y=0.23, xmin=0, xmax=0.5, c='green', lw=1, linestyle='dashed',label ='médiane')
plt.axvline(x=0.5, ymin=0, ymax=0.25, c='green', lw=1, linestyle='dashed')
ax.text(-0.14, 0.23, "0.23",color='green')
ax.text(0.47, -0.10, "0.5",color='green')

plt.xlabel("part cumulée des clients (%)")
plt.ylabel("part cumulée du CA (%)")
plt.legend()
plt.title('Répartition du CA entre les clients\n', fontsize=14)
plt.show()
```

<span style="color:red">**=> 50% des clients génèrent 77% du CA et 50% du CA est généré par 21% des clients**</span>

## ***Lien entre âge et montant total des achats***

```python
# création d'une variable 'age_price' avec les sommes le CA en fonction de l'âge des clients
age_price = lapage.groupby('age').sum().reset_index()
age_price = age_price[['age', 'price']].sort_values(by='age', ascending=False) 
#age_price['price'] = age_price['price']
age_price.head()
```

```python
# visualisation avec un scatterplot
#sns.set_style('ticks')
sns.set_context('talk')
fig, ax = plt.subplots(figsize=(15, 8))
plt.plot(age_price['age'], age_price['price'], 'o', color='red')

plt.xlabel('Age')
plt.ylabel('Montant achat (€)')
plt.title('Montant total des achats selon l\'âge du client', fontsize=20)

plt.show()
```

```python
# discrétisation en 4 groupes d'âge
df = lapage.groupby('age')['price'].sum().reset_index()
df['class_age'] = pd.cut(x=df['age'], bins=[17,30,45,60,94], labels=["18-30","30-45", "45-60", "61+"])
```

```python
X = "class_age" # qualitative
Y = "price" # quantitative

modalites = df[X].unique()
groupes = []
for m in modalites:
    groupes.append(df[df[X]==m][Y])

plt.figure(figsize=(12,8))
medianprops = {'color':"black"}
meanprops = {'marker':'o', 'markeredgecolor':'black',
           'markerfacecolor':'firebrick'}
plt.boxplot(groupes, labels=modalites, showfliers=False, medianprops=medianprops, 
            vert=False, patch_artist=True, showmeans=True, meanprops=meanprops)
plt.title("Montant total des achats selon la classe d'âge du client", fontsize=20)
plt.xlabel("Montant achat (€)")
plt.ylabel("Classe d'age")
plt.grid (True)

plt.show()
```

## ***Lien entre âge et fréquence des achats***

```python
# création d'un nouveau dataFrame
clients_freq = lapage.groupby('age').count().reset_index()
clients_freq = clients_freq[['age', 'session_id']].rename(columns={'session_id': 'nb_tran'})

# création d'une variable fréquence
clients_freq['fréquence'] = clients_freq['nb_tran'] / sum(clients_freq['nb_tran'])
clients_freq.sort_values(by='age', ascending=False).head()
```

```python
# visualisation
plt.figure(figsize = (20, 12))
sns.distplot(clients_freq['fréquence'])
plt.xlabel('Fréquence', fontsize=20)
plt.ylabel('Density', fontsize=20)
plt.title('Distribution des fréquences', y=1.02, fontsize=25)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
#ax.set_ylim(0, 0.05)
plt.grid()
plt.show()
```

```python
#Visualisation avec un scatterplot
fig, ax = plt.subplots(figsize=(15, 8))
plt.plot(clients_freq['age'], clients_freq['fréquence'], 'o')

plt.xlabel('Age')
plt.ylabel('Fréquence achat')
plt.title('Fréquence des achats selon l\'age du client', fontsize=20)

plt.show()
```

### Discrétisation de la variable 'age'

```python
# agrégation des données selon l'âge et la catégorie
age_categ = lapage.groupby(['age', 'categ']).count().reset_index()
age_categ = age_categ[['age', 'categ', 'session_id']]
# création des 3 classes d'âges
cut_bins = [17,30,45,60]
cut_labels = ["18-30","30-45", "45-60"]
age_categ['age']= pd.cut(age_categ['age'], bins=cut_bins, labels=cut_labels)
age_categ = age_categ.groupby(['age','categ']).sum().reset_index()
age_categ.head()
```
