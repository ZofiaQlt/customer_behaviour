## ***Profils clients*** (répartition CA entre eux) Lorenz

```python
# mesure de concentration (courbes de Lorenz et indice de Gini)
depenses = lapage[lapage['price'] > 0]
dep = depenses['price'].values
n = len(dep)
lorenz = np.cumsum(np.sort(dep)) / dep.sum()

# la courbe de Lorenz commence à 0
lorenz = np.append([0],lorenz)

# surface sous la courbe de Lorenz. Le premier segment (lorenz[0]) est à moitié en dessous de 0, on le coupe donc en 2, on fait de même pour le dernier segment lorenz[-1] qui est à moitié au dessus de 1.
AUC = (lorenz.sum() - lorenz[-1] / 2 - lorenz[0] / 2) / n

# surface entre la première bissectrice et le courbe de Lorenz
S = 0.5 - AUC 
gini = 2*S
print(bold + red + f"\nL'indice de Gini est égal à {round(float(gini), 2)}, la répartition des clients par rapport aux CA n'est pas égalitaire")

# il y a un segment de taille n pour chaque individu, plus 1 segment supplémentaire d'ordonnée 0. Le premier segment commence à 0-1/n, et le dernier termine à 1+1/n.
xaxis = np.linspace(0-1/n,1+1/n,n+1)

# visualisation
fig, ax = plt.subplots(figsize=(6, 6))

plt.plot(xaxis,lorenz,drawstyle='steps-post', label = 'courbe de Lorenz')
plt.plot(xaxis,xaxis, label ='égalité parfaite')

plt.axhline(y=0.5, xmin=0, xmax=0.74, c='red', lw=1, linestyle='dashed',label ='médiale')
plt.axvline(x=0.76, ymin=0, ymax=0.5, c='red', lw=1, linestyle='dashed')
ax.text(-0.12, 0.49, "0.5",color='red')
ax.text(0.69, -0.10, "0.76",color='red')

plt.axhline(y=0.25, xmin=0, xmax=0.5, c='green', lw=1, linestyle='dashed',label ='médiane')
plt.axvline(x=0.5, ymin=0, ymax=0.27, c='green', lw=1, linestyle='dashed')
ax.text(-0.14, 0.24, "0.25",color='green')
ax.text(0.47, -0.10, "0.5",color='green')

plt.xlabel("part cumulée des clients (%)")
plt.ylabel("part cumulée du CA (%)")
plt.legend()
plt.title('Répartition du CA entre les clients\n', fontsize=14)
plt.show()
```

<span style="color:red">**=> 50% des clients génèrent 75% du CA et 50% du CA est généré par 24% des clients**</span>

### Comportement des clients

```python
# nombre d'achats par client
client = lapage.groupby('client_id').count().reset_index()
client['nb_tran'] = client['session_id']
client = client[['client_id', 'nb_tran']]
client = client.sort_values(by='nb_tran', ascending=False)
client.head(10)
```

```python
# aperçu des statistiques descriptives du nombre de transactions 
client.describe()
```

<span style="color:red">**=> On remarque que 4 clients se détachent en terme de nombre de transactions d'achats, il doit s'agir de clients B2B :**</span>
+ client c_1609
+ client c_6714
+ client c_3454
+ client c_4958

<span style="color:red">**=> Nous allons le confirmer avec un z-score**</span>

```python
z = stats.zscore(client['nb_tran'])
outlier_z = client[z > 1.96]
nb_outlier_z = len(outlier_z)
print(bold + red + f"\n{nb_outlier_z} produits sont concernés par un z-score > 1,96 soit {round(float(nb_outlier_z / len(client)) * 100, 2)} % des observations :" + end)
outlier_z.head()
```

<span style="color:red">**=> Nous retrouvons bien nos 4 clients B2B**</span>

<span style="color:red">**=> Pour éviter tout biais nous allons dissocier les clients B2B des clients B2C, et les analyser dans deux notebooks séparés**</span>

```python
# autre écriture possible : client_b2b = lapage[(lapage['client_id'] =='c_1609') | (lapage['client_id'] =='c_4958') | (lapage['client_id']=='c_6714') | (lapage['client_id']=='c_3454')].index
values = ['c_1609', 'c_6714', 'c_3454', 'c_4958']
client_b2b = lapage[lapage['client_id'].isin(values)]
# enregistrement du fichier client_b2b
client_b2b.to_csv('src/client_b2b.csv', index=False)

# autre écriture possible : client_b2c = lapage.drop(client_b2b , inplace=True)
client_b2c = lapage[~lapage['client_id'].isin(values)]
# enregistrement du fichier client_b2c
client_b2c.to_csv('src/client_b2c.csv', index=False)

print("\nNombre d'observations avant modification :", len(lapage), "\n\nAprès modification :\n")
print("Nombre d'observations B2B :", len(client_b2b))
print("Nombre d'observations B2C :", len(client_b2c))
```
